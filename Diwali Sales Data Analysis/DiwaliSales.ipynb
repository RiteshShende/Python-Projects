```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
ds = pd.read_csv("Diwali Sales Data.csv", encoding = "unicode escape")
ds.head()
```

```python
ds.info()
```

```python
ds.drop(["Status","unnamed1"],axis = 1, inplace = True)
```

```python
ds.shape
```

```python
ds.isnull().sum()
```

```python
ds.dropna(inplace = True)
ds
```

```python
ds.info()
```

```python
ds["Amount"] = ds["Amount"].astype(int)
```

```python
ds.info()
```

```python
ds.isnull().sum()
```

```python
ds.duplicated().sum()
```

```python
ds.drop_duplicates(inplace = True)
```

```python
ds.duplicated().sum()
```

```python
ds.columns
```

## Exploratory Data Analysis

## Gender

```python
gc = sb.countplot(data = ds, x = "Gender",color="skyblue")
for bars in gc.containers:
    gc.bar_label(bars)
plt.show()
plt.title("Gender distribution")
```

```python
ds["Gender"].value_counts().plot(kind = 'bar', color = "red")
plt.xlabel("Gender",fontsize = 15)
plt.ylabel('Counts', fontsize = 15)
plt.bar_label(bars)
plt.title("Gender distribution")
plt.show()
```

```python
ga = ds.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by = "Amount", ascending=False)
gas = sb.barplot(data = ga,x = "Gender", y = "Amount",palette={"M" : "red", "F" : "blue"})
for bars in gas.containers:
    gas.bar_label(bars)
plt.title("Amount spent by Gender")
plt.legend()
plt.show()
```

# age

```python
ac = sb.countplot(data = ds, x = "Age Group", hue = "Gender")
for bars in ac.containers:
    ac.bar_label(bars)
plt.show()
```

```python
aa = ds.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values("Amount",ascending=False)
sb.barplot(data = aa, x = "Age Group", y = "Amount")
plt.show()
```

```python
ds.columns
```

## State wise order

```python
so = ds.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by = "Orders",ascending=False).head(10)
sb.set(rc = {"figure.figsize":(16,5)})
ss = sb.barplot(x = "State", y = "Orders", data = so, color = "black")
for bars in ss.containers:
    ss.bar_label(bars)
plt.title("State Wise Orders")    
plt.show()
```

## State wise Amount

```python
sa = ds.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by = "Amount",ascending=False).head(10)
sb.set(rc = {"figure.figsize":(16,5)})
ss = sb.barplot(x = "State", y = "Amount", data = sa, color = "black")
for bars in ss.containers:
    ss.bar_label(bars)
plt.title("State Wise Amount")    
plt.show()
```

```python
ds.columns
```

## Amount by Marital Status and gender

```python
ms=sb.countplot(x = "Marital_Status", data = ds, palette= { "0": "pink", "1": "yellow"})
for bars in ms.containers:
    ms.bar_label(bars)
plt.title("Marital Status Distribution")
plt.show()
```

```python
msga = ds.groupby(["Marital_Status", "Gender"], as_index=False)["Amount"].sum().sort_values(by = "Amount", ascending=False)
plt.figure(figsize=(6,5))
sb.barplot(data = msga, x = "Marital_Status", y = "Amount", hue = "Gender")
plt.title("Amount calculation by gender and Marital Status") 
plt.show()
```

## Amount spend by occupation

```python
OA = ds.groupby(["Occupation"], as_index=False)["Amount"].sum().sort_values(by = "Amount", ascending=False)
sb.set(rc = {"figure.figsize" : (20,5)})
sb.barplot(data = OA, x = "Occupation", y = "Amount",palette="Set2")
plt.title("Amount spent by Occupation", fontsize =15 )
plt.show()
```

## Amount spent by Product_Category

```python
CA = ds.groupby(["Product_Category"],as_index= False)["Amount"].sum().sort_values(by = "Amount",ascending=False).head(10)
sb.set(rc = {"figure.figsize":(20,5)})
sb.barplot(x = "Product_Category",y = "Amount", data = CA,palette="Set2")
plt.title("Amount by Category")
plt.show()
```
