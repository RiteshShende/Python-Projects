```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
az = pd.read_csv('amazon.csv', encoding = 'unicode_escape')
az.head()
```

```python
az.info()
```

```python
az.isnull().sum()*100/az.shape[0]
```

```python
az.dropna(inplace = True)
```

```python
az.duplicated().sum()
```

```python
az.shape
```

```python
az.describe()
```

```python
az['m_category'] = az['category'].apply(lambda x:x.split('|')[0])
```

```python
az['m_category'].value_counts()
```

```python
az['discounted_price'] = az['discounted_price'].str.replace(r'[^0-9.]','',regex = True)
```

```python
az['actual_price'] = az['actual_price'].str.replace(r'[^0-9.]','',regex = True)
```

```python
az.head()
```

```python
az.info()
```

```python
az['discounted_price'] = az['discounted_price'].astype('float')
az['actual_price'] = az['actual_price'].astype('float')
```

```python
az['rating'] = az['rating'].astype('float')
az['rating_count'] = az['rating_count'].astype('float')
```

```python
az['rating'].unique()
```

```python
az['rating'].str.contains('\|').sum()
```

```python
az = az[az['rating']!='|']
```

```python
az['rating'].unique()
```

```python
az['rating'] = az['rating'].astype('float')
```

```python
az['rating_count'] = az['rating_count'].str.replace(r'[^0-9.]','',regex = True)
```

```python
az['rating_count'] = az['rating_count'].astype('float')
```

```python
az['discount_percentage'] = az['discount_percentage'].str.replace(r'[^0-9.]','',regex = True)
```

```python
az['discount_percentage'] = az['discount_percentage'].astype('int')
```

```python
az.info()
```

```python
az.describe()
```

```python
az.head()
```

```python
sb.set(rc = {'figure.figsize' : (15,5)})
plt.figure(figsize=(15,5))
ax  = sb.countplot(x = 'm_category', data = az, palette= 'Set2')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
az['revenue'] = az['actual_price']-az['discounted_price']
```

```python
cx = az.groupby(['m_category'],as_index= False)['revenue'].sum().sort_values(by = 'revenue', ascending= False)
sb.barplot(x = 'm_category', y = 'revenue', data = cx)
plt.show()
```
